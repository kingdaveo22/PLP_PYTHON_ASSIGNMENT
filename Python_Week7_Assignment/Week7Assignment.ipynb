```python
# Check the data types
print(df.dtypes)

# Check for missing values
print(df.isnull().sum())
```

```python
# Basic statistics
print(df.describe())

# Grouping by species and computing the mean
grouped = df.groupby('species').mean()
print(grouped)
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal length (cm)', data=df)
plt.title('Average Sepal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['sepal length (cm)'], kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

```python
## Findings and Observations

1. **Bar Chart**: It is observed that the average sepal length varies across species.
2. **Histogram**: The sepal length follows a roughly normal distribution.
3. **Scatter Plot**: There is a positive correlation between sepal length and petal length, with clear differentiation between species.
```
